```python
import pandas as pd
import numpy as np

imdb_list = pd.read_csv('imdb.csv')  # Read the csv file.
imdb_list_df = pd.DataFrame(imdb_list)  # Convert csv into Data Frame.
imdb_list_df = imdb_list_df[['Film_Name','Film_Date','Film_Country','Film_Rate','Director','Stars']]  # Rearrange columns.
imdb_list_sorted = imdb_list_df.sort_values(by="Film_Rate",ascending=False)  # Sort the list by Film_Rate.
imdb_list_sorted.index = np.arange(1,(len(imdb_list_sorted.Film_Name))+1) # Rearrange the index numbers.
imdb_list_sorted
```
